```python
#### Kruskal-Wallace

results = {}

# Iterate over each protein
for protein in long_format['Target'].unique():
    # Filter the DataFrame for the current protein
    protein_df = long_format[long_format['Target'] == protein]

    # Group data by drug and collect values in a list for each drug
    groups = [group['Value'].dropna().values for name, group in protein_df.groupby('Drug')]

    # Perform Kruskal-Wallis H-test if there are more than one group with data
    if len(groups) > 1:
        stat, p = kruskal(*groups)
        results[protein] = {'Statistic': stat, 'P-value': p}

# Print or process the results as needed
for protein, result in results.items():
    print(f"{protein}: Statistic={result['Statistic']}, P-value={result['P-value']}")
```

```python
#Dunn's test ... p-value < 0.05 means significant difference 

# Iterate over each protein
for protein in long_format['Target'].unique():
    # Filter the DataFrame for the current protein
    protein_df = long_format[long_format['Target'] == protein]

    # Perform Kruskal-Wallis H-test
    groups = [group['Value'].dropna().values for name, group in protein_df.groupby('Drug')]
    if len(groups) > 1:
        stat, p = kruskal(*groups)

        # If the Kruskal-Wallis test is significant, perform Dunn's test
        if p < 0.05:
            # Conduct Dunn's test
            dunn_df = protein_df.dropna(subset=['Value'])  # Dropping NaN values
            dunn_test = posthoc_dunn(dunn_df, val_col='Value', group_col='Drug')

            # Output the results
            print(f"Dunn's test for {protein}:\n{dunn_test}\n")
```

```python
# Function to translate p-value to asterisks
def p_value_to_asterisks(p):
    if p < 0.001:
        return '***'
    elif p < 0.01:
        return '**'
    elif p < 0.05:
        return '*'
    else:
        return 'ns'  # not significant

# Iterate over each protein
for protein in long_format['Target'].unique():
    # Filter the DataFrame for the current protein
    protein_df = long_format[long_format['Target'] == protein]

    # Perform Kruskal-Wallis H-test
    groups = [group['Value'].dropna().values for name, group in protein_df.groupby('Drug')]
    if len(groups) > 1:
        stat, p = kruskal(*groups)

        # If the Kruskal-Wallis test is significant, perform Dunn's test
        if p < 0.05:
            # Conduct Dunn's test
            dunn_df = protein_df.dropna(subset=['Value'])
            dunn_test = posthoc_dunn(dunn_df, val_col='Value', group_col='Drug')

            # Translate p-values to asterisks
            significance_matrix = dunn_test.applymap(p_value_to_asterisks)

            # Output the results
            print(f"Dunn's test for {protein}:\n{significance_matrix}\n")
```
